```r
pop <- data.frame("birth"=rep(0,1000), "death"=rep(NA,1000), "smurfness"=rep(0,1000))
params <- list("a" = 0.015,
               "b" = -0.09,
               "k" = 0.19)
```

```r
death_event <- mk_event_individual(
    type = "death",
    intensity_code = '
        if (I.smurfness==0)
            result = 0;
        else
            result = k;
    '
)
```

```r
swap_event <- mk_event_individual(
    type = "swap",
    intensity_code = '
        if (I.smurfness==0)
            result = a*age(I,t)+b;
        else
            result = 0;
    ',
    kernel_code = 'I.smurfness=1'
)
```

```r
model <- mk_model(
    characteristics = get_characteristics(pop),
    events = list(death_event, swap_event),
    parameters = params)
```
